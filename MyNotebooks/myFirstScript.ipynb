# Read file from github as a raw file and write it to a location in google drive

```
import pandas as pd
ChicagoData = pd.read_csv('https://raw.githubusercontent.com/menonpg/covid19_Chicago/master/timeSeries_COVID19_Chicago.csv')
ChicagoData.to_csv('/content/drive/My Drive/MyNotebooks/test.csv', index=False)
```

# Summarize above data

```
print("Min values by column \n" + str(ChicagoData.aggregate('min')) + "\n")
print("Min values by column \n" + str(ChicagoData.aggregate('max')))
```
